## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

# assume VIP are the top 5% customers , preferred customers are the top 25% (besides VIPs)
perc_95 = np.percentile(perCustomer.amount_spent, 95)
perc_75 = np.percentile(perCustomer.amount_spent, 75)
print(perc_75,perc_95)
VIP = perCustomer[perCustomer['amount_spent']>=perc_95]
VIP
VIP.groupby(['Country'], as_index = False)['CustomerID'].count()
```

```python
# with pd.qcut

bins = pd.qcut(perCustomer['amount_spent'], 
               q=[0, .5, .75, .95, 1], 
               labels=['Light', 'Regular', 'Preffered', 'VIP']).to_frame()
perCustomer['label']=bins
perCustomer
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
preferred = perCustomer[perCustomer['amount_spent']>=perc_75]

preferred.groupby(['Country'], as_index = False)['CustomerID'].count()
```
