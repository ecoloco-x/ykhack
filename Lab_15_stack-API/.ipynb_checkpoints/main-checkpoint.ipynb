#### Question 1: Find the questions and answers of last month.

```python
# your code here

# define timeframe
today = datetime.date.today()
yesterday =  today - datetime.timedelta(days=1)
last_month_end = today.replace(day = 1) - datetime.timedelta(days = 1)
last_month_start = last_month_end.replace(day = 1)

print(today)
print(yesterday)
print(last_month_end)
print(last_month_start)
```

```python
# fetch questions and answers from last month
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=last_month_start, todate=last_month_end)
answers = SITE.fetch('answers', fromdate=last_month_start, todate=last_month_end)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here

special_questions = SITE.fetch('questions',
                               fromdate = yesterday, 
                               todate = today, 
                               min = 5, 
                               tagged = 'python', 
                               sort = 'votes', 
                               order = 'desc')
special_questions
```
