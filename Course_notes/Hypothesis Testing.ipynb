## Central limit theorem

The central limit theorem states that the distribution of the sum of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the shape of the original distribution. This means that if you have a large enough sample size, the mean of that sample will be approximately normally distributed, regardless of the distribution of the original population.


In the example below, we generate 1000 random numbers from a gamma distribution, which is not a normal distribution. Then, we calculate the mean of 100 random samples of size 10 from the data. Finally, we plot a histogram of the sample means. As you can see, the histogram of the sample means is approximately normal, demonstrating the central limit theorem in action.

```python
import numpy as np
import matplotlib.pyplot as plt

# generate 1000 random numbers from a non-normal distribution
np.random.seed(0)
data = np.random.gamma(2, 2, 1000)

plt.hist(data)
```

```python
# calculate the mean of 100 random samples of size 10 from the non-normal data
sample_means = [np.mean(np.random.choice(data, 10)) for i in range(100)]

# plot the sample means
plt.hist(sample_means)
plt.show()
```

## Confidence Interval

A confidence interval is a range of values that is expected to contain the true value of a population parameter with a certain level of confidence. For example, a 95% confidence interval for the mean of a population would mean that if we were to repeat the sampling process many times, the true mean of the population would fall within the calculated confidence interval 95% of the time.

Here's an example in Python to illustrate how to calculate a confidence interval for the mean of a population:

```python
import numpy as np
import scipy.stats as stats

# generate a sample of 100 random numbers from a normal distribution
np.random.seed(0)
data = np.random.normal(5, 2, 100)

# calculate the mean and standard deviation of the sample
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# calculate the 95% confidence interval for the mean
alpha = 0.05
n = len(data)
z = stats.norm.ppf(1 - alpha / 2)
lower, upper = sample_mean - z * sample_std / np.sqrt(n), sample_mean + z * sample_std / np.sqrt(n)
print(f"The 95% confidence interval for the mean is: [{lower:.2f}, {upper:.2f}]")
```

In the example above, we generate a sample of 100 random numbers from a normal distribution, and calculate the mean and standard deviation of the sample. Then, we use the `scipy.stats.norm.ppf` function to calculate the critical value for a given significance level (in this case, `alpha=0.05 for a 95% confidence interval`). Finally, we use the critical value, the sample mean, and the sample standard deviation to calculate the lower and upper bounds of the confidence interval, and print the result.

## Standard Normal distribution

The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is also known as the standard normal or the unit normal distribution. In probability and statistics, the standard normal distribution is often used as a reference distribution to compare other normal distributions to.

`Converting a normal distribution` with a given mean and standard deviation to a standard normal distribution involves subtracting the mean and dividing by the standard deviation. The resulting distribution will have a mean of 0 and a standard deviation of 1.

In the example below, we generate 1000 random numbers from a normal distribution with mean 5 and standard deviation 2. Then, we subtract the mean and divide by the standard deviation to convert the normal distribution to a standard normal distribution. Finally, we plot a histogram of the standardized data, and plot the theoretical density function for the standard normal distribution (calculated using the scipy.stats.norm.pdf function) on top of it. As you can see, the histogram of the standardized data is approximately normally distributed with a mean of 0 and a standard deviation of 1, demonstrating the successful conversion to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# generate 1000 random numbers from a normal distribution with mean 5 and standard deviation 2
np.random.seed(0)
data = np.random.normal(5, 2, 1000)

plt.hist(data, bins = 30)
plt.show()
# convert the normal distribution to a standard normal distribution
standardized_data = (data - np.mean(data)) / np.std(data)

# plot the histogram of the standardized data
plt.hist(standardized_data, bins=30, density=True)

# plot the theoretical density function for the standard normal distribution
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, 'r', lw=2)

plt.show()
```

# Hypothesis testing

## Key Concepts

<b>Population</b> - In Statistics, a population is a set of similar items or events which is of interest for some question or experiment. For example, say we want to analyze the income data of all residents in New York City. In this case the population is all the NYC residents.

<b>Sample</b> - A sample is a subset of the population chosen to represent the population in statistical analysis. For example, in reality it is impossible to obtain all residents' income information in NYC because there are too many people. In order to analyze NYC residents' income data, we can select a decent number of residents and use their income information to represent all NYC residents. Ideally, the selection of the sample should be randomized in order to avoid sampling bias.

<b>Sampling Bias</b> - Sampling bias occurs when some members of the intended population are less likely to be included than others in the selected sample. Say we collected the data at Wall Street at 10AM on Monday. Significant bias can be introduced due to the flaws of the data collection because people who appear at Wall Street at that specific time are most likely working professionals in the financial industry. The sample data will miss a lot of people from other industries and other NYC neighborhoods. Consequently, the selected sample is not a good representation of NYC residents.

<b>Statistical Hypothesis</b> - A statistical hypothesis is an assumption about the population average. For example, we can assume the average income of NYC residents is $850,000. This assumption may or may not be true.

<b>Null Hypothesis</b> - Denoted with H0, a null hypothesis is an assumption that the population average is identical to a specific value.  The typical notation is μ = μ0, where μ refers to the population mean and μ0 refers to the hypothesized value. For example, a null hypothesis can be that the average NYC residents' income is $85,000. The null hypothesis is represented by:

H0: μ = 85,000

<b>Alternative Hypothesis</b> - An alternative hypothesis is the opposite of the null hypothesis. We compare this hypothesis with the null hypothesis to decide whether or not we reject the null hypothesis. We denote the alternative hypothesis with H1 or Ha. For the previous example of the null hypothesis, the alternative hypothesis can be any of the following:

H1: μ > 85,000

H1: μ < 85,000

H1: μ ≠ 85,000

## Types of Alternative Hypotheses


There are two types of alternative hypotheses: `one-sided` and `two-sided`. One-sided alternative hypothesis (a.k.a. directional hypothesis) is used to determine whether the population average differs from the hypothesized value in a specific direction (larger but not smaller than, or vice versa). In contrast, two-sided alternative hypothesis (a.k.a. nondirectional hypothesis) is used to determine whether the population average is either greater than or less than the hypothesized value.

Why we differentiate these two types? Because sometimes we are not interested in whether the population average is larger or smaller than the hypothesized value. We only care if they are different. In this case, we use two-sided instead of one-sided hypothesis. However, in case we care about which is the case, we use one-sided hypothesis.

One-sided Alternative Hypothesis

H1: μ > k or H1: μ < k

H1: μ ≠ k

## Elements of Hypothesis testing 

The elements of a hypothesis test include:

1. Null hypothesis (H0): A statement about the population that is assumed to be true until proven otherwise. The null hypothesis is usually a statement of no effect, no difference, or no relationship.


2. Alternative hypothesis (Ha): A statement about the population that contradicts or negates the null hypothesis. The alternative hypothesis is usually a statement of an effect, a difference, or a relationship.


3. Significance level (α): The level of significance, usually expressed as a probability, that is used to determine whether to reject or fail to reject the null hypothesis. A common significance level is 0.05, meaning that there is a 5% chance of making a Type I error (rejecting a true null hypothesis).


4. Test statistic: A numerical value that summarizes the sample data in a way that allows us to test the null hypothesis. The choice of test statistic will depend on the specific problem and data at hand, but common examples include the t-statistic, Z-statistic, and chi-squared statistic.


5. P-value: The probability of observing a test statistic as extreme as the one calculated from the sample data, assuming that the null hypothesis is true. The p-value is used to make a decision about whether to reject or fail to reject the null hypothesis.


6. Critical value: A threshold based on the significance level and the distribution of the test statistic under the null hypothesis. The critical value separates the region in which the null hypothesis is accepted from the region in which it is rejected.


7. Decision: The conclusion about whether to reject or fail to reject the null hypothesis, based on the test statistic and the p-value or the critical value.


8. Conclusion: The interpretation of the results, including the implications for the population and any limitations or assumptions of the hypothesis test.

These elements are used together to form a structured approach to testing a claim about a population based on a sample of data. By conducting a hypothesis test, we can make inferences about the population and determine whether our evidence supports or contradicts the null hypothesis.

## Hypothesis testing steps

the steps for conducting a hypothesis test using test statistics without using a p-value:

1. Formulate the null hypothesis and alternative hypothesis: The null hypothesis is a statement about the population that we hope to prove false, and the alternative hypothesis is a statement about the population that we hope to prove true.

2. Select a significance level: This is the probability of making a Type I error (rejecting a true null hypothesis), typically represented as alpha (α). A common significance level is 0.05.

3. Determine the test statistic: The test statistic is a numerical value that summarizes the sample data in a way that allows us to test the null hypothesis. The choice of test statistic will depend on the specific problem and data at hand, but common examples include the t-statistic, Z-statistic, and chi-squared statistic.

4. Determine the critical value: The critical value is a threshold based on the significance level and the distribution of the test statistic under the null hypothesis. The critical value separates the region in which the null hypothesis is accepted from the region in which it is rejected.

5. Make a decision: Based on the test statistic and the critical value, we either reject or fail to reject the null hypothesis. If the test statistic is greater than or equal to the critical value, we reject the null hypothesis. If the test statistic is less than the critical value, we fail to reject the null hypothesis.

6. Interpret the results: If we reject the null hypothesis, we conclude that there is sufficient evidence to support the alternative hypothesis. If we fail to reject the null hypothesis, we do not have sufficient evidence to support the alternative hypothesis.

In hypothesis testing using test statistics, the goal is to determine whether the test statistic is consistent with the null hypothesis, or whether it is sufficiently different from what we would expect under the null hypothesis to suggest that the alternative hypothesis is true. By comparing the test statistic to the critical value, we can make a decision about whether to reject or fail to reject the null hypothesis, and draw inferences about the population based on the sample data.

## using Z-Test and T-test

A Z-test and a T-test are both statistical tests used to compare the mean of a sample to a population mean. The choice between a Z-test and a T-test depends on the availability of information about the population standard deviation and the sample size.

The Z-test is used when the population standard deviation is known and the sample size is large (n ≥ 30). The test statistic in a Z-test is calculated as:

Z = (x̄ - μ) / (σ / √n)

Where x̄ is the sample mean, μ is the population mean, σ is the population standard deviation, and n is the sample size.

The T-test is used when the population standard deviation is unknown and the sample size is small (n < 30). The test statistic in a T-test is calculated as:

T = (x̄ - μ) / (s / √n)

Where x̄ is the sample mean, μ is the population mean, s is the sample standard deviation, and n is the sample size.

In general, it is preferable to use a Z-test when the population standard deviation is known because the Z-distribution is more reliable and has a more normal distribution than the T-distribution. However, in practice, the population standard deviation is often unknown, and the T-test provides a good approximation when the sample size is small.

## Hypothesis Testing using Z-Test

### Two sided Z-test 

Here is an example of a two-sided Z-test hypothesis test:

A quality control manager is interested in determining whether the mean weight of bags of a new product is equal to the target weight of 10 pounds. The manager takes a sample of 30 bags and calculates the sample mean weight to be 9.95 pounds. The population standard deviation is known to be 0.5 pounds.

Step 1: Formulate the null hypothesis and alternative hypothesis:

H0: μ = 10 (The mean weight of the bags is equal to 10 pounds)

Ha: μ ≠ 10 (The mean weight of the bags is not equal to 10 pounds)

Step 2: Select a significance level:

α = 0.05

Step 3: Determine the test statistic:

Since the population standard deviation is known, we can use the Z-distribution to calculate the test statistic:

Z = (x̄ - μ) / (σ / √n) = (9.95 - 10) / (0.5 / √30) = -1.73

Step 4: Determine the critical value:

Since this is a two-sided test, we need to find the critical values that correspond to the significance level of 0.05 on both sides of the normal distribution. For a two-sided test at α = 0.05, the critical values are ±1.96.

Step 5: Make a decision:

Since the Z-statistic of -1.73 is less than the lower critical value of -1.96, we fail to reject the null hypothesis.

Step 6: Interpret the results:

We do not have sufficient evidence to suggest that the mean weight of the bags is not equal to 10 pounds. The calculated Z-statistic of -1.73 falls within the region of acceptance for the null hypothesis, and the sample mean weight of 9.95 is not significantly different from the target weight of 10 pounds.

This example shows how a Z-test can be used to test a hypothesis about the population mean when the population standard deviation is known. By following the steps of the hypothesis test, we can make an informed decision about whether the sample data supports or contradicts the null hypothesis.

### One sided Z-test

Here is an example of a one-sided Z-test with a known population standard deviation:

A manufacturer is interested in determining whether the average lifespan of a new type of battery is greater than the industry standard of 500 hours. The manufacturer takes a sample of 50 batteries and calculates the sample mean lifespan to be 505 hours. The population standard deviation is known to be 10 hours.

Step 1: Formulate the null hypothesis and alternative hypothesis:

H0: μ = 500 (The average lifespan of the battery is equal to 500 hours)

Ha: μ > 500 (The average lifespan of the battery is greater than 500 hours)

Step 2: Select a significance level:

α = 0.05

Step 3: Determine the test statistic:

Since the population standard deviation is known, we can use the Z-distribution to calculate the test statistic:

Z = (x̄ - μ) / (σ / √n) = (505 - 500) / (10 / √50) = 5

Step 4: Determine the critical value:

Since this is a one-sided test (Ha: μ > 500), we only need to find the critical value that corresponds to the significance level of 0.05 on the upper side of the normal distribution. For a one-sided test at α = 0.05, the critical value is 1.64.

Step 5: Make a decision:

Since the Z-statistic of 5 is greater than the critical value of 1.64, we reject the null hypothesis.

Step 6: Interpret the results:

We have sufficient evidence to suggest that the average lifespan of the battery is greater than 500 hours. The calculated Z-statistic of 5 falls in the region of rejection for the null hypothesis, and the sample mean lifespan of 505 hours is significantly greater than the industry standard of 500 hours.

This example shows how a Z-test can be used to test a hypothesis about the population mean when the population standard deviation is known. By following the steps of the hypothesis test, we can make an informed decision about whether the sample data supports or contradicts the null hypothesis.

## Hypothesis Testing using T-Test


### Two Sided Test 

1. Define the null and alternative hypotheses:
A researcher wants to determine if the average height of adult men in a certain population is different from the average height of adult men in the general population. The null hypothesis (H0) states that the average height of adult men in the population is equal to the average height of adult men in the general population. The alternative hypothesis (Ha) states that the average height of adult men in the population is not equal to the average height of adult men in the general population.

H0: μ = μ0

Ha: μ ≠ μ0

2. Select a significance level (α):
The significance level is set to 0.05, which means that there is a 5% chance of rejecting the null hypothesis when it is actually true.

3. Collect the data:
A sample of 50 adult men from the population is collected and their heights are recorded. The average height of adult men in the general population is known to be 68 inches.

4. Calculate the sample mean (x̄) and sample standard deviation (s):
The mean height for the sample is calculated as x̄ = 67 inches and the standard deviation is calculated as s = 2 inches.

5. Calculate the degrees of freedom (df):
The degrees of freedom can be calculated as df = n - 1, where n is the sample size. In this case, df = 50 - 1 = 49.

6. Calculate the T-statistic:
The T-statistic is calculated as t = (x̄ - μ0) / (s / √n), where μ0 is the population mean specified in the null hypothesis (in this case, μ0 = 68 inches).

7. Find the critical value (tcritical) from a T-distribution table:
The critical value can be found from a T-distribution table with the appropriate degrees of freedom (in this case, df = 49) and the specified significance level (α = 0.05).

8. Compare the T-statistic and critical value:
If the absolute value of the T-statistic is greater than the critical value, then we reject the null hypothesis.

9. Conclude:
Based on the results of the T-test, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest that the average height of adult men in the population is different from the average height of adult men in the general population.

## T-Test using Python

In the example below, the function ttest_1samp performs a one-sample T-test to compare the mean of the sample sample to a population mean of 12. The output of the function is the T-statistic and the p-value. If the p-value is less than the significance level (e.g., 0.05), then we reject the null hypothesis and conclude that the mean of the sample is not equal to the population mean. If the p-value is greater than or equal to the significance level, then we fail to reject the null hypothesis and conclude that the mean of the sample is equal to the population mean.

Here's an example of how to perform a two-tailed T-test in Python using the scipy library's "ttest_1samp" function:

```python
import numpy as np
from scipy.stats import ttest_1samp

# Generate a sample of random data with a mean of 10 and a standard deviation of 2
np.random.seed(0)
sample = np.random.normal(10, 2, 100)

# Perform a two-tailed T-test to compare the mean of the sample to a population mean of 12
t_statistic, p_value = ttest_1samp(sample, 12)

# The null hypothesis is that the mean of the sample is equal to the population mean.
# If the p-value is less than the significance level (e.g., 0.05), then we reject the null hypothesis.
# If the p-value is greater than or equal to the significance level, then we fail to reject the null hypothesis.

if p_value < 0.05:
    print("Reject the null hypothesis. The mean of the sample is not equal to the population mean.")
else:
    print("Fail to reject the null hypothesis. The mean of the sample is equal to the population mean.")
```
